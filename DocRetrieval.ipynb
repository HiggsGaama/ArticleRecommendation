<h1>3. Feature Extraction</h1>
<h3>Generating word count vectors</h3>

```python
#Extracting only the useful features - title, text. More features can be added to the model later on.
newData = data[['title', 'text', 'link']]
print(newData)
```

```python
#writing a function that finds words counts from a given text and returns  dictionary of word-count pairs

def retWordCounts(text):
    vectorizer = CountVectorizer()
    X = vectorizer.fit_transform(text)

    counts = dict(zip(vectorizer.get_feature_names(), list(X.toarray()[0]))) #nd now we have a word-count dictonary
    return counts



#retWordCounts([newData.iloc[0].text])[0] #printing only the word-count dictonary
    
```

```python
newData['wordCounts'] = newData.apply(lambda row: retWordCounts([row.text]), axis = 1) #Only storing the wordCount dictionary in the dataframe
print(newData)
```

<h1>4. Computing TF-IDF for the entire corpus </h1>

```python
#We used the count vector produced earlier to create a word-count tictionary and added it to the table for visualization.
#However, as we need it again to calculate TF-IDF, let's calculate it again
cv = CountVectorizer()
X = cv.fit_transform(newData.text)
print(X.shape)
```

```python
tfTransformer = TfidfTransformer()
tfidf = tfTransformer.fit_transform(X)
print(tfidf.shape)
```

```python
words = cv.get_feature_names()
tfidfVals = tfidf.toarray()
```

```python
print(tfidfVals)
```

<h1 >4.1 A faster way using Pipelines<h6>(faster to code)</h6></h1>

```python
from sklearn.pipeline import Pipeline
import numpy as np
```

```python
pipe = Pipeline([('count', CountVectorizer()),
                ('tfidf', TfidfTransformer())
                ])
```

```python
newData
```

```python
pipe.fit(newData.text)
```

```python
countFromPipe = pipe['count'].transform(newData.text).toarray()
print(countFromPipe)

```

```python
pipeTfidf = pipe['tfidf'].transform(countFromPipe).toarray()
print(pipeTfidf)

del pipeTfidf
del countFromPipe
```

```python
pipeFinal = pipe.transform(newData.text).toarray()
print(pipeFinal)
```

```python
finalTfidfsAndWords= list(pipeFinal.copy())
words = cv.get_feature_names()

for i in range(0, len(finalTfidfsAndWords)):
    a = list(zip(words, finalTfidfsAndWords[i]))
    finalTfidfsAndWords[i] =   list(sorted(a, key = lambda x: x[1], reverse= True))
    
del a
```

<h1>6. Building a k-means clustering model to find 'k' nearest documents</h1>

```python
from sklearn.neighbors import NearestNeighbors
```

```python
neigh = NearestNeighbors()
neigh.fit(pipeFinal)
```

```python
#10 articles similar to article 0 - "Chatbots were the next big thing: what happened? ‚Äì The Startup ‚Äì Medium"
indices = neigh.kneighbors(n_neighbors= 10, X = pipeFinal[0].reshape(1,-1))[1]
newData.iloc[indices[0]]

#Notice how the titles are same for some of the articles. As we found out during data exploration,
#they're all part of a series of articles.
```
